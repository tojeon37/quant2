```python
#결측치가 있는 행 모두 삭제 
df.dropna()
```

```python
#결측치가 있는 열을 삭제
df.dropna(subset=['age'], axis = 0)
```

```python
#결측치가 있는 행을 삭제
df.dropna(axis=1)
```

```python
df.info()
```

```python
#열에 null값이 300개 이상인 열을 삭제하라
df.dropna(axis=1, thresh=300)
```

```python
#결측치가 있는 데이터를 삭제해버리면, 데이터양이 줄어들고 편향되고 분석방향이 엉뚱하게된다
#결측이 있을경우 다른값으로 대체할수 있다.
#결측이 많은경우 결측칸에 나머지 값들의 평균을 넣어 대체 해보자.
df_2 = df.copy() #데이터프레임 복사
df_2.head(6) #상위 6개값
```

```python
#age의 평균으로 nan값을 대체해보자
mean_age = df_2['age'].mean()
mean_age
```

```python
#age 평균값이 29.69이니 결측값에 평균나이를 넣어라, 
# 그리고 inplace = True를 통해 데이터프레임 원본값을 변경해주어라
df_2['age'].fillna(mean_age,inplace=True) 
df_2['age'].head(10)
```

서로 이웃하고 있는 데이터끼리는 유사성을 가질 가능성이 높으며 
시계열데이터는 더욱 그러함. 결측치에는 바로 앞이나 뒤의 값으려 변경하는 것이 좋을수있음
