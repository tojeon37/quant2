데이터를 그림으로 나타낼 경우 직관적으로 이해가가능
파이썬에서는 matplotlib 와 pandas 패키지를 사용하며,
고급시각화에는 seaborn 패키지를 사용한다.


```python
import seaborn as sns
import matplotlib.pyplot as plt


df=sns.load_dataset('penguins')
df.head()
```

```python
#scatter는 산점도그래프, x축에 flipper, y축에 body 데이터를 넣어  그래프를 그리겠다
plt.scatter(df['flipper_length_mm'], df['body_mass_g'])
```

```python
#각 펭귄의 종별 몸무게 차이를 그래프로 그려보자.
#종별로 그룹을 묶어서 몸무게 평균을 구해보자 

df.groupby('species')['body_mass_g'].mean()
```

```python
#그림그릴땐 x,y축이 필요하여, 여기있는 인덱스부분을 초기화 해준다.

df_group = df.groupby('species')['body_mass_g'].mean().reset_index()
plt.bar(df_group['species'], height=df_group['body_mass_g'])
```

```python
#몸무게 히스토그램을 그려보자

plt.hist(df['body_mass_g'])
```

```python
#너무 듬성듬성하니 막대기를 30개로 나눠 그려보자
plt.hist(df['body_mass_g'], bins=30)
```

```python
plt.hist(df['body_mass_g'], bins=30)
plt.xlabel('몸무게') #x축에 라벨설명
plt.ylabel('개채수') #y축에 설명
plt.title('펭귄의 몸무게 분포') #타이틀에 제목설정, 한글로하니 깨져서 영어로 해보자.
```

```python
plt.hist(df['body_mass_g'], bins=30)
plt.xlabel('mass') #x축에 라벨설명
plt.ylabel('count') #y축에 설명
plt.title('mass_count') #타이틀에 제목설정, 한글로하니 깨져서 영어로 해보자.
```

```python
plt.rc('font', family= "Malgun Gothic") # matplotlib에서 한글지원안해줘 깨지니 한글폰트설정
plt.hist(df['body_mass_g'], bins=30)
plt.xlabel('몸무게') #x축에 라벨설명
plt.ylabel('개채수') #y축에 설명
plt.title('펭귄의 몸무게 분포') #타이틀에 제목설정, 한글로하니 깨져서 영어로 해보자.
```
