1.인코딩: 한글로된 데이터를 크롤링하면 깨져보임 : 한글 인코딩으로 발생하는문제
인코딩 : 한글 ---> 0101100 
디코딩 : 0101100 --> 한글
한글을 인코딩하는데 쓰이는 EUC-KR, CP949, UTF-8 정도 이해하도 충분
euc-kr은 한글 문자 2350개에 번호를 붙인방법, 모든자음모음을 표현하기 부족
cp949는 11,720개 한글 문자에 번호를 붙인방법 : 윈도우의 기본인코딩
utf-8은 조합형, 한글과 다양한 언어에 적용, 전세계웹이 대부분 utf-8로만들어짐
