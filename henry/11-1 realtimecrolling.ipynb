셀레니움이란 다양한 브라우저 및 플랫폼에서 웹 응용 프로그램을 테스트할 수 있게 해주는 라이브러리

웹자동화 테스트 용도로 개발되어 실제 브라우저를 사용. 



```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
#driver = webdriver.Chrome(ChromeDriverManager().install())
#위의 1번 driver 명령이 오류시 2번설정으로 한다.


#띄우고 싶은 url을 입력하면 창이 띄워진다.
url = 'https://www.naver.com'
driver.get(url)
```

```python
#html소스를 보여주어라
driver.page_source
```

```python

#브라우저 상에서 보이는 버튼, 검색창, 사진, 테이블, 동영상 등을 element 혹은 요소라 함
#naver의 뉴스창을 클릭한다.
driver.find_element(By.LINK_TEXT, value='뉴스').click()
```

```python
#뒤로가기 실행
driver.back()
```

```python
#검색창에서 검색하기 f12 누르고 검색창 우클릭 "검사" 누르면 f12화면에서 관련된곳이 뜬다.

driver.find_element(By.CLASS_NAME, value ='input_text')
```
