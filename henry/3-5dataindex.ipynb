```python
# 필터링이란 시리즈나 데이터프레임의 데이터에서 조건을 만족하는 원소만 추출하는 것 
# 엑셀의 필터와 비슷, 필터링에는 불리언 인덱싱(boolean indexing)과 isin()메서드가 사용

#시리즈 객체에 조건을 입력하면 각 원소에 대해 참 혹은 거짓을 판별하여 true/false로 이루어진 시리즈가 반환
# 그후 참에 해당하는 데이터만 선택하면 조건을 만족하는 데이터만 추출할수 있음

#실린더스 열에서 고유한 값을 추출해보기 (차의 실린더 갯수 추출)
df['cylinders'].unique()
```

```python
#실린더가 4인 열을 추출해보자
df['cylinders'] == 4
```

```python
# 실린더가 4개인 조건에 해당하는 시리즈를 만들어준다.

filter_bool = (df['cylinders']==4)
df.loc[filter_bool, ]

# 만들어진 불리언 시리즈를 데이터프레임에 입력하면 필터링 할수 있다.
```

```python
#조건이 여러개인 데이터선택, 실린더갯수가4개, 마력이 100인 데이터선택해보기

filter_bool2 = (df['cylinders'] ==4) & (df['horsepower'] >=100)
df.loc[filter_bool2, ]
```

```python
#보고싶은 열만 볼수 있게 출력
df.loc[filter_bool2, ['cylinders','horsepower','name'] ]
```

```python
#isin()메서드를 사용하면 특정 값을 가진 행을 추출할수 있다
filter_isin = df['name'].isin(['ford maverick','for mustang ii','chevrolet impala'])
df.loc[filter_isin]
```
